```python
# to create a new column with the values of the existing columns
data['Full_name']=data['First_Name']+' '+ data['Last_Name']
```

```python
# to create a new dataframe with new values 
perform_data={"Employee_ID":[101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
              "Performance_Rating": [4, 3, 5, 4, 2, 5, 4, 3, 5, 4]}
df_performance_data=pd.DataFrame(perform_data)
df_performance_data
```

```python
# merge performance rating dataframe to the base table data
df_merged=data.merge(df_performance_data, on='Employee_ID',how='left')
df_merged
```

```python
#TO SORT Salary column

df_sorted=data.sort_values(by='Salary', ascending=False)
print(df_sorted)
```

```python
high_salary=data['Salary']>60000
print(high_salary)
```

```python
#to filter certain rows in the column 'Salary'
high_salary=data[data['Salary']>60000]
print(high_salary)
```

```python
# to select multiple columns and creata new dataframe called emplo_info
df_emplo_info=data[['First_Name','Last_Name','Salary']]
print(df_emplo_info)
```

```python
#to convert the date to standard format and to extract year and month
data['Join_Date']=pd.to_datetime(data['Join_Date'])
data['Join_Year']=data['Join_Date'].dt.year
data['Join_month']=data['Join_Date'].dt.month
data
```

```python
current_year=pd.to_datetime('today').year
data['Yearsofexp']=current_year-data['Join_Year']
data
```

```python
df_grouped=data.groupby('Department').agg({'Salary':['mean','median'], 'Yearsofexp':'max'})
df_grouped
```
